```python
bookings['total_kids'] = bookings.children + bookings.babies
bookings.groupby('hotel').aggregate({'total_kids' : 'mean'}).round(2)
```

```python
def has_kids(lst):
    for i in lst:
        if i>0:
            i=True
        else:
            i=False
    return lst

bookings['has_kids'] = has_kids(bookings.total_kids)
```

```python
percents = bookings.query("is_canceled==1").groupby("has_kids",as_index = False).aggregate({"children":"count"})
```
